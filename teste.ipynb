```python
import pandas as pd

data = pd.read_csv('arxiv_data.csv')
```

```python
data.head()
```

```python
import re 
import inflect

inflect = inflect.engine()

stop_words = open("stop_words.txt","r")
stop_words = stop_words.read()  
stop_words = stop_words.split("\n")
stop_words = [word.upper() for word in stop_words]

dici = dict()
for summary in data["summaries"]:
    match = re.findall(r"(\b\w+\b)",summary)
    match = [string.upper() for string in match]
    for word in match:
        if word not in stop_words:
            if word in dici:
                dici[word] += 1
            else:
                dici[word] = 1

for word in dici.keys():
    if inflect.singular_noun(word) and word[-1] == 'S' :    
        if inflect.singular_noun(word) in dici:        
            dici[inflect.singular_noun(word)] = dici[inflect.singular_noun(word)] + dici[word]
            

dici = {k: v for k, v in sorted(dici.items(), key=lambda item: item[1], reverse=True)}
print(dici)
        


```
